```python
import numpy as np
import matplotlib.pyplot as plt
```

$$
    x^2y'' - 4xy' + 6y = 0
$$

```python
x_, x = np.linspace(-5, 0, 100), np.linspace(0, 5, 100)
A = (3, 90)
B = (-4, -112)
plt.scatter(*A, c='red')
plt.scatter(*B, c='red')

C2s = [1, 6, 12]

for C2 in C2s:
    C1 = (A[1] - A[0]**3*C2) / A[0]**2
    C3 = (B[1] - B[0]**2*C1) / B[0]**3
    y1 = lambda t: C1*t**2 + C2*t**3
    y2 = lambda t: C1*t**2 + C3*t**3
    line = plt.plot(x_, y2(x_), label=rf'$C_2 = {C2}$')[0]
    line = plt.plot(x, y1(x), c=line.get_color())[0]

plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
```

```python
x_, x = np.linspace(-4.5, 0, 100), np.linspace(0, 4, 100)
A = (3, 90)
B = (2, 60)
C = (-4, -112)
plt.scatter(*A, c='red')
plt.scatter(*B, c='red')
plt.scatter(*C, c='red')

C2 = (B[1] - B[0]**2 * A[1] / A[0]**2) / (B[0]**3 - B[0]**2 * A[0]**3 / A[0]**2)
C1 = (A[1] - A[0]**3*C2) / A[0]**2
C3 = (C[1] - C[0]**2*C1) / C[0]**3
y1 = lambda t: C1*t**2 + C2*t**3
y2 = lambda t: C1*t**2 + C3*t**3
line = plt.plot(x_, y2(x_))[0]
line = plt.plot(x, y1(x), c=line.get_color())[0]

plt.xlabel('x')
plt.ylabel('y')
plt.grid()
```
