```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

$$
    x^2y'' - 4xy' + 6y = 0
$$

```python
x_, x = np.linspace(-5, 0, 100), np.linspace(0, 5, 100)
A = (3, 90)
B = (-4, -112)
plt.scatter(*A, c='red')
plt.scatter(*B, c='red')

C2s = [1, 6, 12]

for C2 in C2s:
    C1 = (A[1] - A[0]**3*C2) / A[0]**2
    C3 = (B[1] - B[0]**2*C1) / B[0]**3
    y1 = lambda t: C1*t**2 + C2*t**3
    y2 = lambda t: C1*t**2 + C3*t**3
    line = plt.plot(x_, y2(x_), label=rf'$C_2 = {C2}$')[0]
    line = plt.plot(x, y1(x), c=line.get_color())[0]

plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
```

$$
    y = C_k \sin^2 x
$$

```python
t = np.linspace(0.01, 16, 1000)
lim_f = lambda t: np.exp(-0.1 * (t // np.pi) * np.pi)

plt.plot(t,  (np.sin(t) ** 2) * lim_f(t))
plt.plot(t, -(np.sin(t) ** 2) * lim_f(t))
plt.plot(t,  (np.sin(t) ** 2) * signal.square(t) * 0.5 * lim_f(t))
# plt.plot(t, lim_f(t))

plt.xlim(-0.5, 15)
plt.ylim(-3, 3)
plt.grid(axis='both')

plt.axhline(y=0, color='gray')
plt.axvline(x=0, color='gray')
```
